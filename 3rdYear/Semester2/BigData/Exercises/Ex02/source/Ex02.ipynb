```python
# Create SparkSession and sparkcontext
from pyspark.sql import SparkSession
spark = SparkSession.builder\
                    .master("local")\
                    .appName('Ex02')\
                    .getOrCreate()
sc=spark.sparkContext
```

```python
df =spark.read.option("delimiter", ",").option("header", True).csv("input.csv")
df.show()
```

### 4.1. Bài 1: Thực hiện chương trình thống kê số lượng người bình chọn ở mỗi mức.

```python
rating_count = df.groupBy("RATING").count().sort(("RATING"))
rating_count.show()
```

### 4.2. Bài 2: Sắp xếp các phim theo số lượt bình chọn.

```python
movies_count = df.groupBy("MOVIEID").count().sort(("MOVIEID"))
movies_count.show()
```

### 4.3. Bài 3: Thống kê mỗi từ xuất hiện trong tài liệu cho trước: 

a. Trường hợp phân biệt hoa thường

```python
text_file = sc.textFile("wallOfText.txt")
counts = text_file.flatMap(lambda line: line.split(" ")) \
                            .map(lambda word: (word, 1)) \
                           .reduceByKey(lambda x, y: x + y)
```

```python
output1 = counts.collect()
for (word, count) in output1:
    print("%s: %i" % (word, count))
```

b. Trường hợp không phân biệt hoa thường 

```python
counts_nocap = text_file.flatMap(lambda line: line.split(" ")) \
                            .map(lambda word: (word.lower(), 1)) \
                           .reduceByKey(lambda x, y: x + y)
```

```python
output2 = counts_nocap.collect()
for (word, count) in output2:
    print("%s: %i" % (word, count))
```
